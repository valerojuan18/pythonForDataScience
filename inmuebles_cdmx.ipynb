## **Tratamiento de valores nulos**

```python
df = df.fillna(0)
df.isnull().sum()
```

## **Remover registros inconsistentes**

```python
df.query('Valor == 0 | Condominio == 0')
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
df_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
df.drop(df_remover, axis=0, inplace=True)
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
df.Tipo.unique()
```

```python
df.drop('Tipo', axis=1, inplace=True)
df.sample(10)
```

## **Creando filtros**

##Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.##

```python
seleccion1 = df['Habitaciones'] == 1
seleccion1
```

```python
type(seleccion1)
```

```python
df[seleccion1]
```

```python
seleccion2 = df['Valor'] < 4200
df[seleccion2]
```

```python
df[seleccion2].max()
```

```python
filtro1 = (seleccion1) & (seleccion2)
df[filtro1]
```

```python
df1 = df[filtro1]
```

##Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².##

```python
filtro2 = (df['Habitaciones'] >= 2) & (df['Valor'] < 10500) & (df['Area'] > 70)
df[filtro2]
```

```python
df2 = df[filtro2]
```
