## **PROBLEMA DE NEGOCIO**

Se trata de un algoritmo capaz de analizar las caracteristicas de una propiedad, como comodidades, tamaño, ocupación del espacio en un periodo determiando y sugerir al anfitrion un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

### **1. Entendiendo el problema**

```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

### **2. Datos numéricos**

```python
datos_normalizados = pd.json_normalize(datos['info_inmuebles'])
datos_normalizados.head()
```

```python
datos_normalizados.info()
```

```python
columnas = list(datos_normalizados.columns)
columnas
```

```python
datos_extraidos = datos_normalizados.explode(columnas[3:])
datos_extraidos
```

```python
datos_extraidos.reset_index(inplace=True, drop=True)
datos_extraidos.head()
```

```python
datos_extraidos.info()
```

```python
import numpy as np
```

```python
datos_extraidos['max_hospedes'] = datos_extraidos['max_hospedes'].astype(np.int64)
```

```python
datos_extraidos.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos_extraidos[col_numericas] = datos_extraidos[col_numericas].astype(np.int64)
```

```python
datos_extraidos.info()
```

```python
datos_extraidos['evaluacion_general'] = datos_extraidos['evaluacion_general'].astype(np.float64)
```

```python
datos_extraidos.info()
```

```python
# Para remover caracteres de una columna...
datos_extraidos['precio'] = datos_extraidos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos_extraidos.info()
```

```python
datos_extraidos['precio'] = datos_extraidos['precio'].astype(np.float64)
```

```python
datos_extraidos.info()
```

```python
datos_extraidos[['cuota_deposito', 'cuota_limpieza']]
```

```python
# Para remover caracteres en mas de una columna se usa applymap
# (cuando hay mas de una columna deja de ser una serie y pasa a ser un dataframe)
datos_extraidos[['cuota_deposito', 'cuota_limpieza']] = datos_extraidos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos_extraidos.info()
```

```python
datos_extraidos[['cuota_deposito', 'cuota_limpieza']] = datos_extraidos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```
