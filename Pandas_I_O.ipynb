## **Leyendo archivos JSON**

### **Importando archivos JSON**

```python
import pandas as pd
```

```python
archivo_json = 'pacientes.json'
datos_pacientes = pd.read_json(archivo_json)
datos_pacientes
```

```python
datos_pacientes.shape
```

```python
archivo_json_2 = 'pacientes_2.json'
datos_pacientes_2 = pd.read_json(archivo_json_2)
datos_pacientes_2
```

### **Normalizando archivos JSON**

```python
df_normalizado = pd.json_normalize(datos_pacientes_2['Pacientes'])
df_normalizado
```

### **Escribiendo archivos JSON**

```python
df_normalizado.to_json('historico_pacientes_norm.json')
```
